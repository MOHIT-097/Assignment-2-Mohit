# **Que:- 1 Discuss string slicing and provide examples.**

**Ans:-**
String slicing is a way to get a part of a string by specifying the start and end positions. Here’s a simple explanation with examples:

**What is String Slicing?**

String slicing involves taking a portion of a string based on the positions you provide. Think of it like cutting a piece from a string.

**How Does It Work?**

In most programming languages, you use brackets with start and end indices to slice a string. The syntax is usually something like this: string[start:end]. Here’s what the parts mean:

start is the position where the slice begins (inclusive).
end is the position where the slice ends (exclusive).
Examples

**1.Basic Example**

If you have a string text = "Hello, World!" and you want to get the word "Hello", you can slice it like this:

```python
text = "Hello, World!"
slice = text[0:5]  # This takes characters from index 0 to 4
print(slice)  # Output: "Hello"
```

**2.Skipping Characters**

You can also specify a step to skip characters. For example, to get every second character:

```python
text = "Hello, World!"
slice = text[::2]  # This takes every second character
print(slice)  # Output: "Hlo ol!"
```

**3.Omitting Start or End**

If you omit the start index, it defaults to the beginning of the string. If you omit the end index, it defaults to the end of the string.

```python
text = "Hello, World!"
slice1 = text[:5]  # Takes from the start to index 4
slice2 = text[7:]  # Takes from index 7 to the end
print(slice1)  # Output: "Hello"
print(slice2)  # Output: "World!"
```

**4.Negative Indices**

Negative indices count from the end of the string. For example:

```python
text = "Hello, World!"
slice = text[-6:-1]  # Takes characters from index -6 to -2
print(slice)  # Output: "World"
```

In summary, string slicing lets you extract parts of a string by specifying where you want to start and end. It’s a powerful tool for manipulating and analyzing text.




# **Que:-2 Discuss string slicing and provide examples.**

**Ans:-**
String slicing in Python is a method used to extract a portion of a string by specifying a range of indices. It’s a powerful tool for manipulating and analyzing text. Here’s a detailed explanation along with examples:

**1. Basic Syntax:**

The basic syntax for slicing a string is:

**substring = string[start:end]**

Start: The index where the slice begins (inclusive).

end: The index where the slice ends (exclusive).

**2. Examples:**

**Basic Slice**

To get a substring from a string, specify the start and end indices:

```python
text = "Hello, World!"
slice1 = text[0:5]  # Extracts from index 0 to 4
print(slice1)  # Output: "Hello"
```

**Omitting Indices**

If you omit the start index, the slice starts from the beginning of the string. If you omit the end index, the slice goes to the end of the string:

```python
text = "Hello, World!"
slice2 = text[:5]  # Extracts from the beginning to index 4
slice3 = text[7:]  # Extracts from index 7 to the end
print(slice2)  # Output: "Hello"
print(slice3)  # Output: "World!"
```

**Negative Indices**

Negative indices count from the end of the string. -1 refers to the last character:

```python
text = "Hello, World!"
slice4 = text[-6:-1]  # Extracts from the 6th last character to the 2nd last character
print(slice4)  # Output: "World"
```

**Step Parameter**

You can add a step parameter to skip characters:

```python
text = "Hello, World!"
slice5 = text[::2]  # Extracts every second character
print(slice5)  # Output: "Hlo ol!"
```

**Combining Steps with Slicing**

You can combine start, end, and step parameters:

```python
text = "Hello, World!"
slice6 = text[1:10:2]  # Extracts from index 1 to 9, every second character
print(slice6)  # Output: "el,W"
```

**3. Edge Cases:**

1.Start Index Greater than End Index: The result is an empty string.

```python
text = "Hello"
slice7 = text[4:2]  # Start index is greater than end index
print(slice7)  # Output: ""
```

2.Start or End Index Out of Range: Python handles these gracefully.

```python
text = "Hello"
slice8 = text[0:10]  # End index is greater than the string length
print(slice8)  # Output: "Hello"
```

String slicing is a fundamental feature in Python that allows for flexible manipulation of strings, making it easier to extract and work with substrings.

# **Que:-3 Describe how to access, modify, and delete elements in a list with examples.**

**Ans:-**In Python, lists are versatile data structures that allow you to access, modify, and delete elements. Here’s how you can perform these operations:

**1. Accessing Elements**

You can access elements in a list using their index. Python uses zero-based indexing, so the first element is at index 0.

Example:-

```python
my_list = ['apple', 'banana', 'cherry']

# Accessing the first element
print(my_list[0])  # Output: 'apple'

# Accessing the second element
print(my_list[1])  # Output: 'banana'
```

**2. Modifying Elements**

You can change the value of an element in a list by assigning a new value to a specific index.
Example:-

```python
my_list = ['apple', 'banana', 'cherry']

# Modifying the second element
my_list[1] = 'blueberry'
print(my_list)  # Output: ['apple', 'blueberry', 'cherry']
```

**3. Deleting Elements**

You can remove elements from a list using different methods:

Using del: Deletes an element at a specified index.

Example:

```python
my_list = ['apple', 'banana', 'cherry']

# Deleting the second element
del my_list[1]
print(my_list)  # Output: ['apple', 'cherry']
```

**Using pop():**

Removes and returns an element at a specified index. If no index is provided, it removes and returns the last element.

Example:

```python
my_list = ['apple', 'banana', 'cherry']

# Removing and returning the last element
last_element = my_list.pop()
print(last_element)  # Output: 'cherry'
print(my_list)       # Output: ['apple', 'banana']

# Removing and returning the second element
second_element = my_list.pop(1)
print(second_element)  # Output: 'banana'
print(my_list)         # Output: ['apple']
```

**Using remove():**

Removes the first occurrence of a specific value from the list.

Example:

```python
my_list = ['apple', 'banana', 'cherry']

# Removing the element with value 'banana'
my_list.remove('banana')
print(my_list)  # Output: ['apple', 'cherry']
```

These operations enable you to effectively manage the contents of a list, allowing for dynamic and flexible data handling.

# **Que:-5 Describe the key features of sets and provide examples of their use.**

**Ans:-** Sets in Python are a collection data type that offers several unique features. Here’s a description of their key features along with examples of their use:

**1. Unordered Collection**

Feature: Sets do not maintain any order among their elements. The order in which elements are added is not preserved.

Example:

```python
my_set = {3, 1, 2}
print(my_set)  # Output: {1, 2, 3} (The order may vary)
```

**2. Unique Elements
Feature:**

Sets automatically eliminate duplicate elements. Each element in a set must be unique.

Example:

```python
my_set = {1, 2, 2, 3}
print(my_set)  # Output: {1, 2, 3}
```

**3. Mutable
Feature:**

 Sets are mutable, meaning you can add and remove elements after the set is created.

Example:

```python
my_set = {1, 2, 3}
my_set.add(4)  # Adding an element
my_set.remove(2)  # Removing an element
print(my_set)  # Output: {1, 3, 4}
```

**4. No Indexing
Feature:**

Sets do not support indexing or slicing because they are unordered. You cannot access elements by index.

Example:

```python
my_set = {1, 2, 3}
# print(my_set[0])  # This would raise a TypeError
```

**5. Set Operations
Feature:**

Sets support mathematical operations like union, intersection, difference, and symmetric difference.

Examples:

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

union_set = set1.union(set2)  # Union of sets
print(union_set)  # Output: {1, 2, 3, 4, 5}

intersection_set = set1.intersection(set2)  # Intersection of sets
print(intersection_set)  # Output: {3}

difference_set = set1.difference(set2)  # Difference of sets
print(difference_set)  # Output: {1, 2}

symmetric_difference_set = set1.symmetric_difference(set2)  # Symmetric difference of sets
print(symmetric_difference_set)  # Output: {1, 2, 4, 5}
```

**6. Set Comprehensions
Feature:**

 Like list comprehensions, sets support set comprehensions to create new sets from iterables.

Example:

```python
squares = {x**2 for x in range(6)}
print(squares)  # Output: {0, 1, 4, 9, 16, 25}
```

**7. Immutable Sets (Frozenset)
Feature:**

A frozenset is an immutable version of a set. Once created, its elements cannot be modified.

Example:

```python
frozen_set = frozenset([1, 2, 3])
print(frozen_set)  # Output: frozenset({1, 2, 3})
# frozen_set.add(4)  # This would raise an AttributeError
```

In summary, sets in Python provide an efficient way to store unique elements, perform set operations, and leverage comprehensions for generating sets. Their unordered nature and support for set operations make them useful for various applications in data processing and manipulation.

# **Que:-7 Describe how to add, modify, and delete items in a dictionary with examples.**

**Ans:-**
Dictionaries in Python are versatile data structures that store key-value pairs. Here’s how you can add, modify, and delete items in a dictionary, along with examples for each operation:

**1. Adding Items**

You can add new key-value pairs to a dictionary by simply assigning a value to a new key.

Example:

```python
# Initial dictionary
my_dict = {'name': 'Alice', 'age': 30}

# Adding a new item
my_dict['city'] = 'New York'

print(my_dict)
# Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}
```

**2. Modifying Items**

To modify an existing item in a dictionary, you reassign a new value to an existing key.

Example:

```python
# Initial dictionary
my_dict = {'name': 'Alice', 'age': 30}

# Modifying an existing item
my_dict['age'] = 31

print(my_dict)
# Output: {'name': 'Alice', 'age': 31}

```

**3. Deleting Items**

You can delete items from a dictionary using the del statement or the pop() method.

Using del: Deletes an item by specifying its key.

Example:

```python
# Initial dictionary
my_dict = {'name': 'Alice', 'age': 30, 'city': 'New York'}

# Deleting an item
del my_dict['city']

print(my_dict)
# Output: {'name': 'Alice', 'age': 30}
```

**Using pop():**

Removes an item by key and returns its value. If the key is not found, it raises a KeyError unless a default value is provided.

Example:

```python
# Initial dictionary
my_dict = {'name': 'Alice', 'age': 30, 'city': 'New York'}

# Removing an item and getting its value
removed_value = my_dict.pop('city')

print(removed_value)
# Output: 'New York'

print(my_dict)
# Output: {'name': 'Alice', 'age': 30}
```

Example with default value:

```python
# Removing an item with a default value if the key is not found
removed_value = my_dict.pop('country', 'Not Found')

print(removed_value)
# Output: 'Not Found'
```

**Summary**

**Adding Items:** Use dictionary[key] = value to add a new key-value pair.

**Modifying Items:** Reassign a new value to an existing key using dictionary
**[key] = new_value.**
**Deleting Items:** Use del **dictionary[key]** or **dictionary.pop(key)** to remove items from the dictionary.
These operations allow you to effectively manage and manipulate the data stored in dictionaries.

# **Que:- 8 Discuss the importance of dictionary keys being immutable and provide examples.**


In Python, dictionary keys must be immutable, which means their value cannot change once they are set. This immutability is crucial for several reasons:

Importance of Immutable Dictionary Keys

**1.Hashing Consistency**

Explanation: Python dictionaries use a hashing mechanism to quickly access values based on keys. Immutable keys ensure that their hash value remains constant, which allows dictionaries to efficiently locate and retrieve values. If keys were mutable, their hash value could change, causing issues in data retrieval.
Example



```python
my_dict = {}
key = (1, 2)  # Immutable tuple
my_dict[key] = "value"
print(my_dict)  # Output: {(1, 2): 'value'}

# Attempting to use a mutable key like a list will raise an error:
# my_dict[[1, 2]] = "value"  # Raises TypeError: unhashable type: 'list'
```

**2.Data Integrity**

Explanation: Immutability guarantees that once a key is set in a dictionary, it cannot be altered. This helps maintain the integrity of the data structure, ensuring that the relationship between keys and values remains consistent.Example:


```python
my_dict = {'a': 1, 'b': 2}
# Keys 'a' and 'b' are immutable, so their association with values 1 and 2 is stable.
```

**3.Efficient Lookups**

Explanation: Because immutable keys have a consistent hash value, dictionaries can quickly perform operations such as lookups, insertions, and deletions. This efficiency is essential for the performance of data-heavy applications. Example:


```python
my_dict = {42: 'Answer to the Ultimate Question'}
print(my_dict[42])  # Output: 'Answer to the Ultimate Question'
```

**4.Predictable Behavior**

Explanation: When keys are immutable, their behavior is predictable during various operations. The dictionary can reliably perform updates and deletions without concerns about the key's value changing unexpectedly.Example:


```python
my_dict = {10: 'ten', 20: 'twenty'}
# The keys 10 and 20 are immutable integers, so dictionary operations remain predictable.
```

**Examples of Immutable and Mutable Keys**

**1.Immutable Keys (Valid):**

Strings: Strings are immutable and can be used as dictionary keys.

```python
my_dict = {'name': 'Alice'}
print(my_dict['name'])  # Output: 'Alice'
```

**Tuples:** Tuples are immutable and can be used as dictionary keys if they contain only other immutable types.

```python
my_dict = {(1, 2): 'value'}
print(my_dict[(1, 2)])  # Output: 'value'
```

**2.Mutable Keys (Invalid):**

Lists: Lists are mutable and cannot be used as dictionary keys.

```python
my_dict = {}
# my_dict[[1, 2]] = 'value'  # Raises TypeError: unhashable type: 'list'
```

**Dictionaries:** Dictionaries themselves are mutable and cannot be used as keys.

```python
my_dict = {}
# my_dict[{'key': 'value'}] = 'another value'  # Raises TypeError: unhashable type: 'dict'
```

**Summary**

The requirement for dictionary keys to be immutable ensures consistent and efficient data access, integrity, and behavior. Immutable keys guarantee that the dictionary structure remains reliable and performance-efficient, which is fundamental for many applications in Python programming.
